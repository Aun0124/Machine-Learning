```
# data type conversion
priceofcars=cars[cars.price.str.isnumeric()].copy()
priceofcars.price.dtype
```

```
hpcars=priceofcars[priceofcars.horsepower.str.isnumeric()].copy()
hpcars.horsepower=hpcars.horsepower.astype(float)
```

```
hpcars.plot.scatter(x='horsepower',y='city_mpg',color='green')
```

```
pricegroup=priceofcars.groupby(['make'])['price']
```

```
means=pricegroup.mean()
std=pricegroup.std()
```

```
fig,ax=plt.subplots()   #subplots(row ,column ,rowsize,columnsize)
means.plot.bar(x=ax,yerr=std,color='lightgreen')  #ax for  x-axis ,yerr for errorbar(precision)
```
